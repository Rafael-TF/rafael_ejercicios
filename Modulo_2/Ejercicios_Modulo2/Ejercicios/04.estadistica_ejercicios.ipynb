### Ejercicio 01:
- Define una funci칩n que tome como entrada un array y retorne la varianza. Ap칩yate en la siguiente f칩rmula:
$$ \mathbf{S^2} = \frac{1}{n} \sum_{i=1}^{n} (x_{i}^{2}- \overline{x}^2)$$

- Comprueba que funciona correctamente compar치ndola con _**np.var()**_.

```python
fuel_ciudad = np.array(fuel_city)

def calcular_varianza(array):
    # Paso 1: Tama침o del array
    n = len(array)
    
    # Paso 2: Media de los datos
    media = np.mean(array)
    
    # Paso 3: Suma de las diferencias al cuadrado (seg칰n la f칩rmula)
    suma = np.sum(array**2) - n * (media**2)
    
    # Paso 4: Aplicar la f칩rmula
    varianza = suma / n
    
    return varianza
```

```python
varianza_np = np.var(fuel_ciudad)

print(f"La varianza usando np.var() es: {varianza_np}")

print(f"La varianza calculandolo a traves de la funci칩n calcular_varianza es: {calcular_varianza(fuel_ciudad)}")
```

### Ejercicio 02:
- Define una funci칩n que tome como entrada un array y retorne la desviaci칩n est치ndar. Ap칩yate en la siguiente f칩rmula:

$$ \mathbf{S} = \sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} x_{i}^{2}- \overline{x}^2}$$

- Comprueba que funciona correctamente comprob치ndola con _**np.std()**_.

```python
def calcular_desviacion_estandar(array):
    
    # Paso 1: Tama침o del array
    n = len(array)
    
    # Paso 2: Suma de los elementos
    suma_elementos = np.sum(array)
    
    # Paso 3: Suma de los cuadrados
    suma_cuadrados = np.sum(array**2)
    
    # Paso 4: Calcular la varianza usando la f칩rmula correcta
    varianza = (suma_cuadrados / n) - (suma_elementos / n)**2
    
    desviacion_estandar = np.sqrt(varianza)
    
    return desviacion_estandar
    
    
```

```python
desviacion_estandar_numpy = np.std(fuel_ciudad)

print(f"La desviaci칩n estandar usando np.std() es: {desviacion_estandar_numpy}")

print(f"La desviaci칩n estandar a traves de la funci칩n calcular_desviacion_estandar() es: {calcular_desviacion_estandar(fuel_ciudad)}")
```

### Ejercicio 03:
- Define una funci칩n que tome como par치metro un array y retorne los siguientes estad칤sticos en forma de diccionario:
    - M칤nimo
    - M치ximo
    - Media
    - Cuartiles Q1, Q2 (mediana) y Q3
    - Rango intercuartil
    - Desviaci칩n est치ndar
- Prueba la funci칩n con los arrays **fuel_city** y **fuel_hwy**.

```python
def calcular_estadisticos(array):
    
    # C치lculo de los estad칤sticos
    minimo = np.min(array)
    maximo = np.max(array)
    media = np.mean(array)
    mediana = np.median(array)
    q1 = np.percentile(array, 25)
    q3 = np.percentile(array, 75)
    rango_intercuartil = q3 - q1
    desviacion_estandar = np.std(array)
    
    # Diccionario con los resultados
    estadisticos = {
        "M칤nimo": minimo,
        "M치ximo": maximo,
        "Media": media,
        "Mediana (Q2)": mediana,
        "Q1": q1,
        "Q3": q3,
        "Rango Intercuartil": rango_intercuartil,
        "Desviaci칩n Est치ndar": desviacion_estandar
    }
    
    return estadisticos
```

```python
estadisticas_fuel_city= calcular_estadisticos(fuel_ciudad)

fuel_carretera = np.array(fuel_hwy)

estadisticas_fuel_carretera = calcular_estadisticos(fuel_hwy)

print(f"Los datos estad칤sticos de la ciudad son: {estadisticas_fuel_city}")
print(f"Los datos estad칤sticos de la carretera son: {estadisticas_fuel_carretera}")
```

### Ejercicio 04:
- Define una funci칩n que toma un array como entrada, y retorna el mismo array sin los outliers.
- Utiliza la **Puntuaci칩n Z** para el filtrado de valores at칤picos.
- Prueba la funci칩n con los arrays _**fuel_city**_ y _**fuel_hwy**_, y calcula qu칠 porcentaje de datos se ha conservado tras el filtrado de outliers para cada caso.

```python
def filtrar_outliers(array, z = 3):
    
    # Calculamos media y std
    mean = np.mean(array)
    std = np.std(array)
    
    # Calculamos los limites laterales (洧랞 췀 z*洧랥)
    lim_l = mean - z*std
    lim_r = mean + z*std
    
    # Filtramos los elementos del array, los que esten fuera de los limites laterales.
    outliers = [elem for elem in array if elem < lim_l or elem > lim_r]
    
    # Filtramos los elementos del array, los que no pertenezcan a outliers.
    normal_data = [elem for elem in array if elem not in outliers]
    
    # retornar array_filtrado
    return np.array(normal_data)
```

```python
fuel_city_filtrado = filtrar_outliers(fuel_ciudad)
fuel_hwy_filtrado = filtrar_outliers(fuel_carretera)

print("Fuel Ciudad:", fuel_city_filtrado)
print("Fuel Carretera:", fuel_hwy_filtrado)
```

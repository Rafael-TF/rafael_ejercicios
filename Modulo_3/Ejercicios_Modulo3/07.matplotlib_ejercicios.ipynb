### Ejercicio 5:
- Obtener un histograma con las columnas _**Largo Petalo**_ y _**Ancho Petalo**_ del dataset de **iris**.

<div align="center">
   
</div>

```python
# Por separado:
plt.figure(figsize=(3,3))
plt.hist(df['Largo Petalo'], alpha = 0.5)
plt.show()

plt.figure(figsize=(3,3))
plt.hist(df['Ancho Petalo'], alpha = 0.5)
plt.show()
```

```python
# superpuestos: ideal si están relacionados o se quieren comparar, y tienen escalas cercanas
plt.hist(df['Largo Petalo'], alpha = 0.5)
plt.hist(df['Ancho Petalo'], alpha = 0.5)
plt.show()
```

```python
# en horizontal
plt.hist(df['Largo Petalo'], alpha = 0.5, orientation='horizontal')
plt.hist(df['Ancho Petalo'], alpha = 0.5, orientation='horizontal')
plt.show()
```

```python
plt.figure(figsize = (8, 4))

# una fila y dos columnas
plt.subplot(1, 2, 1)
plt.hist(df['Largo Petalo'], alpha = 0.5)

plt.subplot(1, 2, 2)
plt.hist(df['Ancho Petalo'], alpha = 0.5)

plt.show()
```

```python
plt.figure(figsize = (4, 4))

# dos filas y una columna
plt.subplot(2, 1, 1)
plt.hist(df['Largo Petalo'], alpha = 0.5)

plt.subplot(2, 1, 2)
plt.hist(df['Ancho Petalo'], alpha = 0.5)

plt.show()
```

### Ejercicio 9:
- Muestra en una gráfica de sectores el número de los que no sobrevivieron por clase (_**Pclass**_).

<div align="center">
   
</div>

```python
# df[df['Survived'] == 0].value_counts('Pclass', sort=False)
df[df['Survived'] == 0].value_counts('Pclass', sort=False, normalize=True)
```

```python
# normalize True nos da las proporciones en vez de los conteos
values = df[df['Survived'] == 0].value_counts('Pclass', sort=False, normalize=True)
# labels= ['Clase 1', 'Clase 2', 'Clase 3']
labels = sorted([f"clase {i}" for i in df['Pclass'].unique()])

plt.pie(x=values, 
        labels=labels, 
        shadow=True, 
        colors=['green', 'red', 'cyan'], 
        explode = (0.1, 0.1, 0.1),
        autopct = "%1.2f%%",
        textprops={'fontsize': 15},
        labeldistance=1.45
        );
```

### Ejercicio 10:
- Muestra en 2 graficas de sectores con _**plt.subplots()**_ la siguiente información:
    1. En la primera gráfica mostrar el número de pasajeros que murieron de cada clase.
    2. En la segunda gráfica mostrar el número de pasajeros que sobrevivieron de cada clase.

<div align="center">
   
</div>

```python
labels = sorted([f"clase {i}" for i in df['Pclass'].unique()])

plt.subplot(1, 2, 1)
values = df[df['Survived'] == 0].value_counts('Pclass', sort=False, normalize=True)
plt.pie(x=values, 
        labels=labels, 
        colors=['green', 'red', 'cyan'], 
        explode = (0.1, 0.1, 0.1),
        autopct = "%1.2f%%")

plt.subplot(1, 2, 2)
values = df[df['Survived'] == 1].value_counts('Pclass', sort=False, normalize=True)
plt.pie(x=values, 
        labels=labels, 
        colors=['green', 'red', 'cyan'], 
        explode = (0.1, 0.1, 0.1),
        autopct = "%1.2f%%")

plt.title("Not survived vs. Survived %");
```

### Ejercicio 2:
- Usando la ecuación de la recta, encuentra la recta de los dos puntos del ejercicio anterior.

$$
y = \frac{(y_2 - y_1)(x - x_1)}{x_2 - x_1} + y_1
$$

- Usando _**x = np.linspace(-10, 10, 2)**_ grafica la recta.
- Grafica también los dos puntos.

```python
P1 = (-2, -2)
P2 = (2, 5)

x = np.linspace(-10, 10, 2)
y = ((P2[1] - P1[1]) * (x - P1[0]) / (P2[0] - P1[0])) + P1[1]

# Creamos el punto 1 (P1)
plt.plot(P1[0], P1[1], marker = "o", color = "red")

# Creamos el punto 2 (P2)
plt.plot(P2[0], P2[1], marker = "o", color = "blue")

# Recta
plt.plot(x, y, color = "orange")

# Eje X
plt.axhline(0, color = "black")

# Eje Y
plt.axvline(0, color = "black")

plt.grid()
plt.show()
```

### Ejercicio 3:

- Define una función que tome como parámetro un número entero $n$ y que genere una lista de números enteros aleatorios entre el 1 y el 10.000 usando la librería _**random**_.

```python
def lista_aleatoria(n):
    
    if n <= 0:
        
        print("El número debe ser superior a 0")
  
    return [random.randint(1, 10000) for i in range(n)]

lista_aleatoria(5)
```

### Ejercicio 4:

- Define una función que tome como parámetro un número entero $n$ y que genere una lista de 0's y 1's de forma aleatoria usando la librería _**random**_.

```python
def generar_binarios_0_1(n):
    
    opciones= [0, 1]
    
    return [random.choice(opciones) for i in range(n)]

generar_binarios_0_1(5)
```

### Ejercicio 5:

- Define una función que reciba dos listas y que calcule la _**distancia Euclideana**_.

$$D(P1, P2) = \sqrt[]{\sum_{i = 0}^{n} (x_{i} - y_{i})^{2}}$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def calcular_distancia_euclidiana(lista_1, lista_2):
    
    distancia = 0
    
    for i in range(len(lista_1)):
        distancia += (lista_1[i] - lista_2[i]) ** 2
        
    return np.sqrt(distancia)
```

```python
def calcular_distancia_euclidiana(lista_1, lista_2):
    
    distancia = 0 
    
    for x, y in zip(lista_1, lista_2):
        
        distancia += (x - y)**2
        
    return np.sqrt(distancia)
```

```python
def calcular_distancia_euclidiana(lista_1, lista_2):
    return np.sqrt(np.sum([(x - y)**2 for x, y in zip(lista_1, lista_2)]))
```

```python
lista_1 = lista_aleatoria(10)
lista_2 = lista_aleatoria(10)

print(lista_1)
print(lista_2)

print(calcular_distancia_euclidiana(lista_1, lista_2))
```

```python
lista_1 = generar_binarios_0_1(10)
lista_2 = generar_binarios_0_1(10)

print(lista_1)
print(lista_2)

print(calcular_distancia_euclidiana(lista_1, lista_2))
```

### Ejercicio 4:

- Define una función que reciba dos listas y que calcule la _**distancia Manhattan**_.

$$\text {Manhattan} = \sum_{i = 0}^{n}\left| x_{i} - y_{i} \right|$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def calcular_distancia_manhattan(lista_1, lista_2):
    
    distancia = 0
    
    for x, y in zip(lista_1, lista_2):
        
        distancia += np.abs(x - y)
    
    return distancia   
```

```python
def calcular_distancia_manhattan(lista_1, lista_2):
    return sum(np.abs(x - y) for x, y in zip(lista_1, lista_2))
```

```python
lista_1 = lista_aleatoria(10)
lista_2 = lista_aleatoria(10)

print(calcular_distancia_manhattan(lista_1, lista_2))
```

```python
lista_1 = generar_binarios_0_1(10)
lista_2 = generar_binarios_0_1(10)

print(calcular_distancia_manhattan(lista_1, lista_2))
```

### Ejercicio 5:

- Define una función que reciba dos listas y que calcule la _**distancia Chebyshev**_.

$$\text{Chebysev} = max(\left| x_{1} - x_{0} \right|, \left| y_{1} - y_{0} \right|, ..., \left| k_{1} - k_{0} \right|)$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def calcular_distancia_chebyshev(lista_1, lista_2):
    return max(np.abs(x - y) for x, y in zip(lista_1, lista_2))
```

```python
lista_1 = lista_aleatoria(10)
lista_2 = lista_aleatoria(10)

print(calcular_distancia_chebyshev(lista_1, lista_2))
```

```python
lista_1 = generar_binarios_0_1(10)
lista_2 = generar_binarios_0_1(10)

print(calcular_distancia_chebyshev(lista_1, lista_2))
```

### Ejercicio 6:

- Define una función que reciba dos listas y un valor $p$ y que calcule la _**distancia Minkowski**_.

$$
\text{Minkowski} = \left(\sum_{i=1}^{n}\left|x_i - y_i\right|^{p}\right)^{\frac{1}{p}}
$$

- Genera dos listas con la función del _**ejercicio 1**_ función y calcula la distancia de esas dos listas.

- Genera dos listas con la función del _**ejercicio 2**_ función y calcula la distancia de esas dos listas.

```python
def calcular_distancia_minkowski(lista_1, lista_2, p):
   # Calcular la suma de las diferencias absolutas elevadas a la potencia p
    suma_potencias = sum(abs(x - y)**p for x, y in zip(lista_1, lista_2))
    
    # Calcular la raíz p-ésima de la suma
    return suma_potencias**(1/p)
```

```python
lista_1 = lista_aleatoria(10)
lista_2 = lista_aleatoria(10)

distancia = calcular_distancia_minkowski(lista_1, lista_2, 5)
print("Distancia de Minkowski:", distancia)
```

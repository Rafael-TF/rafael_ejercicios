Estrategias para el tratamiento de valores nulos:

* Filtrar aquellos que no sean null: 
    * df[~df['Age'].isnull()]

* Eliminar filas con nulos: 
    * Si hay pocos nulos < 2%  se podría plantear directamente borrarlos
    * df.dropna()

* Eliminar columnas con nulos:
    * Para casos que haya muchos nulos > 80 %
    * df.drop("Clase", axis = 1)

* Rellenar los valores nulos:
    * Si es numérico habitualmente se rellena con la media o mediana.
    * df['Age'].fillna(20)
    * Si es texto o categórico: se rellena con la moda o con un valor fijo, por ejemplo "Otros"
    * Si es fecha: se puede rellenar con el valor justo anterior o el posterior
    * Para imputación más sofisticada:
        * Aplicar una función con lógica para decidir cómo rellenar ese valor en base a otras columnas
        * utilizar imputers de scikit learn que hagan predicciones
        * Utilizar aprendizaje no supervisado para etiquetar en clústers

```python
# rellenar con mediana
df['Age'] = df['Age'].fillna(df['Age'].median())
```

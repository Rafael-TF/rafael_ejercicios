```python
import pandas as pd

df = pd.read_csv('../Data/iris.csv')

df.head()
```

```python
df['Clase'].value_counts()
```

```python
mapeo_clase = {
    'Iris-setora' : 1,
    'Iris-versicolor' : 2,
    'Iris-virginica' : 3
}

df['Clase'] = df['Clase'].map(mapeo_clase)
df.head()
```

```python
# Matriz correlaci√≥n

correlacion = df.corr()
correlacion
```

```python
# Mostrar la matriz de correlaci√≥n con Seaborn
import seaborn as sns

sns.heatmap(correlacion, annot=True, cmap = 'viridis')
```

## Encoding con .get_dummies()

```python
df = pd.read_csv('../Data/iris.csv')
```

```python
pd.get_dummies(df)
```

```python
df_encoded = pd.get_dummies(df, columns=['Clase'])
df_encoded
```

```python
correlacion = df_encoded.corr()
sns.heatmap(correlacion, annot=True, cmap = 'viridis')
```

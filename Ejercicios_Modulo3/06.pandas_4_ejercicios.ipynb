#### En la siguiente celda se guarda la información del excel

```python
df1 = pd.read_excel("../Data/Airbnb_NYC.xlsx", sheet_name = "NYC1")
df2 = pd.read_excel("../Data/Airbnb_NYC.xlsx", sheet_name = "NYC2")
```

### Ejercicio 0:
- Une los dos DataFrames anteriores por la columna **id**. Guarda en nuevo DataFrame en **df**

```python
# Cargar los DataFrames (ya asumimos que están cargados como df1 y df2)
df = pd.merge(df1, df2, on='id', how='inner')
```

```python
df.info()
```

### Ejercicio 1:
- Si quiero hospedarme en Manhattan una sola noche y para elegir el Airbnb me guio por el que tenga el mayor número de reviews, ¿donde me debería de hospedar?

```python
# Filtrar los datos para Manhattan
df_manhattan = df[df['municipio'] == 'Manhattan']

# Filtrar para aquellos que tengan minimum_nights igual a 1
df_manhattan_one_night = df_manhattan[df_manhattan['minimum_nights'] == 1]

# Ordenar por número de reviews en orden descendente
df_manhattan_one_night_sorted = df_manhattan_one_night.sort_values(by='number_of_reviews', ascending=False)

# Obtener el primer registro que tenga el mayor número de reseñas
top_airbnb = df_manhattan_one_night_sorted.iloc[0]

# Mostrar la información del Airbnb con más reseñas
top_airbnb_info = top_airbnb[['name', 'host_name', 'price', 'number_of_reviews', 'latitude', 'longitude']]
top_airbnb_info
```

### Ejercicio 2:
- Busco un Airbnb en el municipio donde el precio promedio de alquiler sea el más bajo y la cantidad de review por mes sea el mayor. ¿Cuáles son las 3 habitaciones que mejor se adaptan a estos criterios?

```python
# Paso 1: Calcular el precio promedio por municipio
avg_price_per_municipio = df.groupby('municipio')['price'].mean()

# Paso 2: Encontrar el municipio con el precio promedio más bajo
municipio_con_precio_bajo = avg_price_per_municipio.idxmin()

# Paso 3: Filtrar los datos para obtener solo los Airbnbs en el municipio con el precio más bajo
df_municipio_bajo = df[df['municipio'] == municipio_con_precio_bajo]

# Paso 4: Ordenar las habitaciones en ese municipio por el número de reseñas por mes de mayor a menor
df_municipio_bajo_sorted = df_municipio_bajo.sort_values(by='reviews_per_month', ascending=False)

# Paso 5: Seleccionar las 3 habitaciones con mayor número de reseñas por mes
top_3_airbnb = df_municipio_bajo_sorted.head(3)

# Mostrar las tres habitaciones más adaptadas a los criterios
top_3_airbnb_info = top_3_airbnb[['name', 'host_name', 'price', 'reviews_per_month', 'room_type', 'municipio']]
top_3_airbnb_info
```

### Ejercicio 3:
- Quiero rentar un Airbnb completo por un total de 200 días, en Brooklyn o Manhattan; no quiero pagar más de 150 dólares por noche, ¿Cuáles son las 5 opciones más economicas que me ofrece Airbnb?

```python
# Paso 1: Filtrar por municipio (Brooklyn o Manhattan)
df_filtered = df[df['municipio'].isin(['Brooklyn', 'Manhattan'])]

# Paso 2: Filtrar por tipo de habitación "entera"
df_filtered = df_filtered[df_filtered['room_type'] == 'Entire home/apt']

# Paso 3: Filtrar por precio máximo de 150 dólares por noche
df_filtered = df_filtered[df_filtered['price'] <= 150]

# Paso 4: Calcular el precio total para 200 días
df_filtered['total_price_200_days'] = df_filtered['price'] * 200

# Paso 5: Ordenar las opciones por el precio total de 200 días (de menor a mayor)
df_filtered_sorted = df_filtered.sort_values(by='total_price_200_days', ascending=True)

# Paso 6: Seleccionar las 5 opciones más económicas
top_5_airbnb = df_filtered_sorted.head(5)

# Mostrar las 5 opciones más económicas
top_5_airbnb_info = top_5_airbnb[['name', 'host_name', 'price', 'total_price_200_days', 'room_type', 'municipio']]
top_5_airbnb_info
```

### Ejercicio 4:
- En un nuevo Dataframe muestra por municipio: el número de habitaciones tipo individual, casa o departamento entero y compartido; también muestra en otra columna el precio promedio de todas las habitaciones por municipio.

```python
# Paso 1: Agrupar por municipio y tipo de habitación, luego contar el número de habitaciones de cada tipo
room_counts = df.groupby(['municipio', 'room_type']).size().unstack(fill_value=0)

# Paso 2: Calcular el precio promedio por municipio
price_avg = df.groupby('municipio')['price'].mean()

# Paso 3: Combinar el número de habitaciones con el precio promedio
result = room_counts.join(price_avg, on='municipio', rsuffix='_avg')

# Mostrar el DataFrame final
result
```

### Ejercicio 5:
- Crea una nueva columna en el DataFrame original, con el rating de acuerdo a la siguiente formula, donde las variables son: puntaje del municipio, número de reviews y disponibilidad de la habitación al año.

- Formula: Rating = (Puntaje de ciudad) + (número de review/ 629) + (disponibilidad / 365).

- El puntaje para cada ciudad es el siguiente {"Manhattan" : 10/10, "Brooklyn" : 8/10, "Queens" : 6/10, "Staten Island" : 4/10, "Bronx" : 2/10}.

```python
# Paso 1: Crear una nueva columna 'puntaje_ciudad' usando el diccionario de puntajes
df['puntaje_ciudad'] = df['municipio'].map({
    "Manhattan": 10,
    "Brooklyn": 8,
    "Queens": 6,
    "Staten Island": 4,
    "Bronx": 2
})

# Paso 2: Calcular el rating de manera vectorizada
df['rating'] = (df['puntaje_ciudad'] +
                (df['number_of_reviews'] / 629) +
                (df['availability_365'] / 365))

# Ver las primeras filas para asegurarnos que la columna fue creada correctamente
df[['municipio', 'number_of_reviews', 'availability_365', 'rating']].head()
```

### Ejercicio 6:
- Encuentra las 5 habitaciones de cada municipio con mayor rating.

```python
# Paso 1: Ordenar el DataFrame por municipio y rating en orden descendente
df_sorted = df.sort_values(by=['municipio', 'rating'], ascending=[True, False])

# Paso 2: Obtener las 5 habitaciones con mayor rating para cada municipio
top_5_rooms_per_municipio = df_sorted.groupby('municipio').head(5)

# Mostrar el resultado
top_5_rooms_per_municipio[['municipio', 'name', 'rating']].reset_index(drop=True)
```

### Ejercicio 7:
- Crea un DataFrame con la cantidad de reviews de cada municipio de acuerdo a tipo de habitación.

```python
# Paso 1: Agrupar por municipio y tipo de habitación y calcular la suma de reviews
reviews_by_municipio_room_type = df.groupby(['municipio', 'room_type'])['number_of_reviews'].sum().reset_index()

# Mostrar el DataFrame resultante
reviews_by_municipio_room_type
```

### Ejercicio 8:
- Encuentra los 10 barrios con mayor precio por habitación individual.

```python
# Paso 1: Filtrar las habitaciones de tipo "Private room" (Habitación privada)
habitaciones_privadas = df[df['room_type'] == 'Private room']

# Paso 2: Ordenar por el precio de mayor a menor y seleccionar los 10 primeros barrios
top_10_barrio_precio = habitaciones_privadas[['barrio', 'price']].sort_values(by='price', ascending=False).head(10)

# Mostrar los 10 barrios con mayor precio por habitación privada
top_10_barrio_precio
```

### Ejercicio 9:
- Encuentra el barrio con mejor rating de cada municipio.

```python
# Paso 1: Agrupar por municipio y encontrar el barrio con mejor rating en cada municipio
mejor_barrio_por_municipio = df.loc[df.groupby('municipio')['rating'].idxmax()][['municipio', 'barrio', 'rating']]

# Mostrar el barrio con mejor rating por municipio
mejor_barrio_por_municipio
```

### Ejercicio 10:
- Del resultado anterio, selecciona los 2 barrios con el menor precio por "dep/casa completo".

```python
# Paso 1: Filtrar las habitaciones de tipo "Entire home/apt"
habitaciones_completas = df[df['room_type'] == 'Entire home/apt']

# Paso 2: Agrupar por municipio y ordenar por precio ascendente para cada municipio
# Seleccionar los 2 barrios con el menor precio
top_2_barrio_precio_completo = habitaciones_completas.groupby('municipio').apply(
    lambda x: x.nsmallest(2, 'price')
)[['barrio', 'price']]

# Mostrar los 2 barrios con menor precio por "Entire home/apt" por municipio
top_2_barrio_precio_completo
```

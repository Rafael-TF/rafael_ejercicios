### Ejercicio 26:
- Encontrar el año en que hubo más migración por continente y crear un dataframe con nombre de continente, el año de máxima migración y cantidad de migración.

|     |Continente|Año |Migraciones|
|-----|----------|----|-----------|
|**0**|Asia      |2010|163845     |
|**1**|Europe    |1990|68301      |
|**2**|Africa    |2010|40892      |
|**3**|Oceania   |1992|2871       |
|**4**|America   |1992|44699      |

```python

# Crear una lista de los años (ya que están como enteros, no como cadenas)
años = list(range(1980, 2014))  # Crear lista con los años de 1980 a 2013

# Filtrar el DataFrame para mantener solo la columna 'Continente' y las columnas de los años
df_migracion = df[['Continente'] + años]  # Seleccionamos la columna de Continente y las columnas de los años

# Agrupar por continente y sumar los valores de migración para cada año
df_migracion = df_migracion.groupby('Continente')[años].sum()  # Sumamos por continente

# Crear una lista para almacenar los resultados
resultados = []

# Iterar sobre cada continente para encontrar el año con la mayor migración
for continente in df_migracion.index:
    max_migracion = df_migracion.loc[continente].max()  # Encontramos la migración máxima
    año_max = df_migracion.loc[continente].idxmax()  # Encontramos el año de la migración máxima
    
    # Agregar el resultado a la lista
    resultados.append({
        'Continente': continente,
        'Año': año_max,
        'Migraciones': max_migracion
    })

# Crear el DataFrame de los resultados
df_max_migracion = pd.DataFrame(resultados)

# Mostrar el DataFrame resultante
print(df_max_migracion)
```

#### Extra: Obtener un Dataframe que te parezca relevante a partir de la aplicación de filtros. ¡Se creativo!

```python
"""Crear un DataFrame con los países que han tenido una migración significativa 
(por ejemplo, total de migración superior a un umbral) 
en continentes específicos y durante un rango de años, para identificar tendencias
"""

# Definir filtros
umbral_migracion = 100000  # Por ejemplo, más de 100,000 migrantes en total
continentes_interes = ['Asia', 'Europa']  # Filtrar solo para Asia y Europa
anios_interes = list(range(2000, 2011))  # Rango de años: 2000 a 2010

# Crear una nueva columna con la suma de migraciones en el rango de años seleccionado
df['total_migracion_anios'] = df[anios_interes].sum(axis=1)

# Aplicar filtros
df_filtrado = df[
    (df['total_migracion_anios'] > umbral_migracion) & 
    (df['Continente'].isin(continentes_interes))
]

# Ordenar los resultados para mayor claridad
df_filtrado = df_filtrado.sort_values(by='total_migracion_anios', ascending=False)

# Mostrar las primeras filas del DataFrame filtrado
print(df_filtrado[['Pais', 'Continente', 'total_migracion_anios']])
```
